6.	Jelaskan bagaimana karakter pengelompokan setiap kelas pada scatterplot

**Jawab:**

*   **Setosa**, bunga iris dengan jenis setosa memiliki ukuran **petal** dan **panjang sepal** yang cenderung **lebih kecil**, apabila dibandingkan dengan bunga iris dengan jenis lainnya. Namun, berbeda dengan ukuran lebar dari sepalnya, setosa justru memiliki **lebar sepal** yang hampir sama dengan jenis lainnya. Di beberapa kasus, ada bunga jenis setosa yang memiliki lebar sepal tertinggi sekaligus terendah dari keseluruhan bunga iris yang dapat kita lihat pada persebaran scatterplot.

*   **Versicolor**, apabila dilihat dari perbandingan ukuran sepal dan petalnya, maka karakter pengelompokkan pada kelas versicolor cenderung memiliki penyebaran nilai di area tengah. artinya, bunga iris jenis versicolor memiliki karakteristik ukuran sepal dan petal yang **sedang** apabila dibandingkan dengan jenis bunga iris lainnya.

*   **Virginica**, jenis verginica memiliki ukuran **petal** dan **panjang sepal** yang **besar** daripada bunga iris jenis lainnya. Namun, untuk ukuran **lebar sepal** masih kalah dibanding dengan jenis setosa

*   **Unknown**, ini merupakan kelas yang dibuat pada bunga iris yang belum diketahui jenisnya. karakteristik pada kelas ini, penyebarannya cenderung mendekati ketiga jenis atau kelompok dari bunga iris, baik itu setosa, versicolor maupun verginica.

8.	Buatlah pararrel coordinat plot dari data yang berkelas setosa, versicolor, dan virginica. Berikan warna merah untuk setosa, hijau untuk versicolor, biru untuk virginica. Hint gunakan perintah plot dan operasi slicing pada numpy untuk mempermudah tugas anda 


```python
# array baru untuk menampung data semua jenis bunga tanpa 'unknown'
new_arr = iris_arr[np.where(iris_arr[:, 4] != 'unknown')]
```

```python
new_arr
```

```python
import matplotlib.patches as mpatches

fig = plt.figure(figsize=(8, 5))
label_x = np.array(['Petal_width', 'Petal_length',	'Sepal_width', 'Sepal_length'])
red_patch = mpatches.Patch(color='red', label='setosa')
blue_patch = mpatches.Patch(color='blue', label='verginica')
green_patch = mpatches.Patch(color='green', label='versicolor')

for i in new_arr:
  if i[4] == 'Setosa':
    plt.plot(label_x, i[0:4], color='red')
  elif i[4] == 'Verginica':
    plt.plot(label_x, i[0:4], color='blue')
  elif i[4] == 'Versicolor':
    plt.plot(label_x, i[0:4], color='green')

plt.title('Parallel Coordinates Plot')
plt.legend(handles=[red_patch, blue_patch, green_patch])
plt.show()
```
