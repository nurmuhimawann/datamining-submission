```python
import pandas as pd

df = pd.read_csv('iris.csv')
df_copy = df.copy()
df
```

##### 2. Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max sehingga setiap nilai attribut memiliki range 0 – 1. Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Vbaru = nilai data di suatu sel setelah normalisasi, Vlama = nilai data asli sebelum normalisasi, Vmax = nilai max dari kolom data tersebut, Vmin = nilai min dari kolom data tersebut  

```python
def normalize(columns):
  return (df_copy[columns] - df_copy[columns].min()) / (df_copy[columns].max() - df_copy[columns].min())
```

```python
columns = df.columns[:4]
for i in columns:
  df_copy[i] = normalize(i)

df_normalize = df_copy
df_normalize
```

##### 3. Cari centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata2 seluruh data suatu kolom

```python
def centroid(df_normalize, var):
  centroid = df_normalize[df_normalize["variety"] == var].iloc[:, :4].mean()
  return centroid
```

```python
variety = df_normalize["variety"].unique()
variety
```

```python
center = {}

for i in variety:
  center['centroid_{0}'.format(i)] = centroid(df_normalize, i)
```

```python
centroid = pd.DataFrame(center)
centroid
```

##### 4. Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan vmin dan vmax pada data 1

```python
df_test = pd.read_csv('iristesting.csv')
df_test.head()
```

```python
def normalize_test(columns):
  return (df_test[columns] - df_test[columns].min()) / (df_test[columns].max() - df_test[columns].min())

columns = df_test.columns[:4]
for i in columns:
  df_test[i] = normalize_test(i)

df_normalize_test = df_test
df_normalize_test.head()
```

##### 5. Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance 

```python
def distance(df_normalize_test, centroid):
  distance = 0
  for i in centroid.index:
    distance += (df_normalize_test[i] - centroid[i])**2
  distance_result = distance**(1/2)
  return distance_result

def predict (x, y):
  variety = df_normalize_test["variety"].unique()
  data_sum = len(x[variety[0]])
  result = []

  true = 0
  false = 0
  for i in range(data_sum):
    compare = []
    for j in variety:
      compare.append(x[j][i])
    index_compare = compare.index(min(compare))
    prediction = variety[index_compare]
    result.append(prediction)

    if prediction == y[i]:
      print("True")
      true += 1
    else:
      print("False")
      false += 1
  return {"result": result,
          "correct": true,
          "wrong": false}

setosa_pred = []
versicolor_pred = []
virginica_pred = []
y = df_normalize_test["variety"]

for i, row in df_normalize_test.iterrows():
  setosa = distance(row, center["centroid_Setosa"])
  versicolor = distance(row, center["centroid_Versicolor"])
  virginica = distance(row, center["centroid_Virginica"])

  setosa_pred.append(setosa)
  versicolor_pred.append(versicolor)
  virginica_pred.append(virginica)

x = {
    "Setosa": setosa_pred,
    "Versicolor": versicolor_pred,
    "Virginica": virginica_pred
}

prediction_result = predict(x, y)
```

##### 6. Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class 

```python
correct = prediction_result['correct']
wrong = prediction_result['wrong']
accuracy = correct/(correct + wrong)*100

print(f"Total data    : {correct + wrong}"
      f"\nPrediksi benar: {correct}"
      f"\nPrediksi salah: {wrong}"
      f"\nAccuracy      : {round(accuracy, 2)} %")
```
