### ***Understanding Data with Exploration Data Analysis (EDA) and Preprocessing Data***

```python
data.drop(data[['fnlwgt', 'educational-num']], axis=1, inplace=True)
data
```

here I simply assume that the attribute 'education' alone is sufficient and does not require 'education-num'. then the 'fnlwght' attribute didn't seem to have any effect on the class either, so I decided to remove it.

***limit the dataset only to data that has an age of 25 - 40 years***

```python
data = data[(data['age'] > 24) & (data['age'] < 41)]   # subseting
data.reset_index(drop=True, inplace=True)              # reset id
data
```

***Check for null data***

```python
(data.isnull() | data.empty | data.isna()).sum()
```

Here the output of the program tells that no data is missing or blank. But after I double checked, there is a string '?' which indicates that the data still contains missing data and the value is unknown. so i need to do some data processing.

I would change the string '?' be the NaN values

```python
data[data == '?'] = np.nan
```

```python
(data.isnull() | data.empty | data.isna()).sum()
```

```python
data[(data['workclass'].isna()) & (data['occupation'].isna()) & (data['income'] == '<=50K')]
```

For data of categorical type, it is better to handle missing values using the mode value of the column. However, I assume that the missing values in the 'workclass' and 'occupation' columns are because they don't have work. so their occupation is unknown which causes the missing value or string '?' to occur. this is reinforced by my analysis which shows a column where the missing 'workclass', 'occupation' mostly have income '<= 50K'. which means they may not have any income at all. So, I decided instead of filling in the missing value with the mode, I'd better fill it with the value 'unemployed'. As for the 'native-country' column, I filled it with the mode value.

```python
print(data['workclass'].mode())
print(data['workclass'].unique())
print('========================================================================')
print(data['occupation'].mode())
print(data['occupation'].unique())
```

```python
for col in ['workclass']:
    data[col].fillna('Never-worked', inplace=True)
```

```python
for col in ['occupation', 'native-country']:
    data[col].fillna('Unemployed', inplace=True)
```

```python
for col in ['native-country']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

### **Modelling**

```python
def ModelNaiveBayesClassifier(X, Y, test, random):
  # train test split
  from sklearn.model_selection import train_test_split
  x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = test, random_state = random)

  # NBC
  from sklearn.naive_bayes import GaussianNB
  gb = GaussianNB()
  gb.fit(x_train,y_train)
  
  # printing metric
  from sklearn.metrics import accuracy_score
  y_pred = gb.predict(x_test)
  print(accuracy_score(y_test,y_pred)*100)
```

***Testing data == 20%***

```python
X = data.drop(['income'], axis = 1)
Y = data['income']
ModelNaiveBayesClassifier(X, Y, 0.2, 42)
```
